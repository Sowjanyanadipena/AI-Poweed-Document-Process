```python
# =========================
# 1. Install dependencies
# =========================
!pip install easyocr transformers gradio pdfplumber pytesseract pillow

# =========================
# 2. Import libraries
# =========================
import easyocr
import pdfplumber
import pytesseract
from PIL import Image
from transformers import pipeline
import gradio as gr
import os

# Initialize OCR and NER
ocr_reader = easyocr.Reader(['en'])
ner_pipeline = pipeline("ner", model="dslim/bert-base-NER")

# =========================
# 3. OCR function
# =========================
def extract_text_from_image(image_path):
    result = ocr_reader.readtext(image_path, detail=0)
    return " ".join(result)

def extract_text_from_pdf(pdf_file):
    text = ""
    with pdfplumber.open(pdf_file) as pdf:
        for page in pdf.pages:
            text += page.extract_text() + "\n"
    return text

# =========================
# 4. AI Processing (NER)
# =========================
def process_text_with_ner(text):
    entities = ner_pipeline(text)
    return entities

# =========================
# 5. Combined pipeline
# =========================
def process_document(file):
    filename = file.name
    if filename.endswith(".pdf"):
        extracted_text = extract_text_from_pdf(filename)
    else:
        extracted_text = extract_text_from_image(filename)

    entities = process_text_with_ner(extracted_text)
    return {"Extracted Text": extracted_text, "Entities": entities}

# =========================
# 6. Gradio UI
# =========================
iface = gr.Interface(
    fn=process_document,
    inputs=gr.File(file_types=[".jpg", ".png", ".pdf"], label="Upload Document"),
    outputs=[
        gr.Textbox(label="Extracted Text"),
        gr.JSON(label="Named Entities")
    ],
    title="AI-powered Document Processing System",
    description="Upload an image or PDF document. This tool extracts text using OCR and identifies entities using AI."
)

iface.launch(share=True)
```
